```python
import numpy as np
import os
import cv2
import matplotlib.pyplot as plt
initial_path = '../cleandata/(3)/fail' # 이미지가 들어있는 폴더의 경로. 실제 이미지는 다음과 같이 저장되어 있습니다.
                                       # '../cleandata/(3)/fail/image.png' 여기서 '../ '의 의미는 현재 디렉토리의 상위 경로를 의미하며
                                       # './'의 의미는 현재 디렉토리를 의미합니다. 모두 상대경로를 의미하고 절대 경로를 쓰고 싶다면
                                       # 'C:/data/cannon/image.png' 혹은 'C:\\data\\cannon\\image.png' 이렇게 입력하시면 됩니다.
image_list=os.listdir(initial_path)    # os.listdir('.')은 . 내부의 파일 리스트를 리스트 형태로 저장하는 명령어입니다.
n = np.random.randint(len(image_list)) # np.random.randint(.)는 .의 숫자만큼 랜덤한 정수를 생성하라는 의미입니다.
image_path = os.path.join(initial_path, image_list[n]) # 
image = cv2.imread(image_path) # opencv가 실제로 쓰일 때는 cv2로 쓰입니다.
image_np = np.array(image)     # opencv 의 데이터 형식을 numpy array로 바꿔주었습니다.
image_np = image_np[240:550, 300:600, :] # 원하는 영역을 크롭하였습니다. 순서대로 (x, y, z) 좌표계를 의미하며 x는 이미지의 세로 높이, y는 가로, z는 채널입니다.
print(image_np.shape) # image의 사이즈(dimension)을 출력하라는 의미입니다.
plt.imshow(image_np)
plt.axis('off')
plt.show()
```

```python
import numpy as np
import os
import cv2
from PIL import Image
import matplotlib.pyplot as plt
initial_path = '../cleandata/(3)/pass'
image_list=os.listdir(initial_path)
n = np.random.randint(len(image_list))
image_path = os.path.join(initial_path, image_list[n])
image = Image.open(image_path)
image_np = np.array(image)
image_np = image_np[50:600, 200:700, :]
print(image_np.shape)
plt.imshow(image_np)
plt.axis('off')
plt.show()
```
